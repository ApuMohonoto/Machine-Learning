# Logistic Regression(Multiple Input)


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv(r"R:\CSV Files\placement.csv")
data.drop(columns="apu",inplace=True)
data.head(3)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="iq",data=data,hue="placement")# hue shows is the plot linearly seperable or not
plt.legend(loc=1)# location
plt.show()
# from this grap we can see we can seperate the orange and blue color by a line.so it is linearly seperable
```

```python
x=data.iloc[:,:-1]
y=data["placement"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_train,y_train)
lg.score(x_test,y_test)*100
```

```python
lg.predict([[5.9,106]])
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lg)
plt.show()
```
