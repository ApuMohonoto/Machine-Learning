```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv(r"R:\CSV Files\Ice_cream selling data.csv")
data.head(3)
```

```python
plt.scatter(data["Temperature (°C)"],data["Ice Cream Sales (units)"])
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (units)")
plt.show()
```

```python
data.isnull().sum()
```

```python
data.ndim
```

```python
x=data[["Temperature (°C)"]]
y=data["Ice Cream Sales (units)"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=2)
pf.fit(x)
x=pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
prd=lr.predict(x)
```

```python
plt.scatter(data["Temperature (°C)"],data["Ice Cream Sales (units)"])
plt.plot(data["Temperature (°C)"],prd,c="red")
plt.xlabel("Temperature (°C)")
plt.ylabel("Ice Cream Sales (units)")
plt.legend(["Scatter","predicated"])
plt.show()
```
