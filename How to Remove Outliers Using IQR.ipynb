```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data=pd.read_csv(r"R:\CSV Files\loan.csv")
data.head(4)
```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x="CoapplicantIncome",data=data)
plt.show()
```

```python
# IQR= Inter Quantile Range=Q3-Q1,min=Q1-1.5*IQR,max=Q3+1.5*IQR
q1=data["CoapplicantIncome"].quantile(.25)
q3=data["CoapplicantIncome"].quantile(.75)
q1,q3
```

```python
IQR=q3-q1
min=q1-(1.5*IQR)
max=q3+(1.5*IQR)
min,max
```

```python
new_data=data["CoapplicantIncome"]<=max
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_data)# outliner is out from boxplot
plt.show()
```
