```python
# min-max technique
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv(r"R:\CSV Files\loan.csv")
data.head(3)
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
m=MinMaxScaler()
m.fit(data[["CoapplicantIncome"]])
data["CoapplicantIncome_MinMax"]=m.transform(data[["CoapplicantIncome"]])
data.head(4)
```

```python
plt.figure(figsize=(12,15))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(data["CoapplicantIncome"])
plt.subplot(1,2,2)
plt.title("after")
sns.distplot(data["CoapplicantIncome_MinMax"])
plt.show()
```
