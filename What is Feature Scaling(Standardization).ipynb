```python
# standardization
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data=pd.read_csv(r"R:\CSV Files\loan.csv")
data.head(3)
```

```python
data.isnull().sum()
```

```python
data["ApplicantIncome"].fillna(data["ApplicantIncome"].mean(),inplace=True)
```

```python
sns.distplot(data["ApplicantIncome"])
```

```python
data.describe()
```

```python
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
d=ss.fit_transform(data[['ApplicantIncome']])
data['ApplicantIncome_ss']=pd.DataFrame(d,columns=["x"])

# By feature Scaling the Magnitude of the data is changed but the nature of the data is same 



```

```python
data.head(3)
```

```python
data.describe()
```

```python
sns.distplot(data["ApplicantIncome_ss"])
plt.show()
```

```python
plt.figure(figsize=(12,15))
plt.subplot(1,2,1)
plt.title("After")
sns.distplot(data["ApplicantIncome_ss"])
plt.subplot(1,2,2)
plt.title("Before")
sns.distplot(data["ApplicantIncome"])
plt.show()
```
