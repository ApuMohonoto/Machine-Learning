```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv(r"R:\CSV Files\placement.csv")
data.head(3)
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="package",data=data)# simple linear regression
plt.show()
```

```python
x=data[["cgpa"]]# dependent variable must be in 2 in two dimension
y=data["package"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)# value can be randomly changes so random_state is used
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)# m,c er best value dey
```

```python
lr.coef_ # value 0f m
```

```python
lr.intercept_ # value 0f c
```

```python
lr.score(x_test,y_test)*100 # accuracy determination
```

```python
6.89 	3.26
```

```python
lr.predict([[6.89]]) # predication of cgpa
```

```python
x=data[["cgpa"]]
y_pre=lr.predict(x)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="package",data=data)
plt.plot(data["cgpa"],y_pre,c="red")
plt.legend(["Orginal","predicated_Line"])
plt.savefig("predict.jpg")
plt.show()
```
