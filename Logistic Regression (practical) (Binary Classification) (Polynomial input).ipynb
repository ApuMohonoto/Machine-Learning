# Logistic Regression (Binary Classification)(Polynomial Input)

```python
#Without Polynomial
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv(r"R:\CSV Files\Ice_cream selling data.csv")
data.head(3)
```

```python
sns.scatterplot(x="Temperature (째C)",y="Ice Cream Sales (units)",data=data,hue="output")
```

```python
x=data.iloc[:,:-1]
y=data["output"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)*100
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```

```python
# with Polynomial
x=data.iloc[:,:-1]
y=data["output"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=4)
pf.fit(x)
x=pd.DataFrame(pf.transform(x))
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)*100
```
