```python
# with scikit
import pandas as pd
data=pd.DataFrame({"Name":["s","xl","m","xl","l","m","s"]})
data
```

```python
order=[["s","m","l","xl","xxl"]]
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe=OrdinalEncoder(categories=order)
oe.fit(data[["Name"]])
data["Ordinal_Encoding with scikit"]=oe.transform(data[["Name"]])
data
```

```python
# With map
order1={"s":5,"m":4,"l":2,"xl":0}
data["Ordinal_Encoding with map"]=data["Name"].map(order1)
data
```
